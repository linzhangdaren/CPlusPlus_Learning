```python

# ====================== 初始化列表 ======================
# 列表是Python中最常用的可变序列（元素可增删改）
# 初始化方式1：直接用中括号包裹元素（推荐）
list1 = [1, 2, 3]  
# 初始化方式2：使用list()函数（空列表或转换其他可迭代对象）
list2 = list()  # 空列表: []
str_list = list("abc")  # 字符串转列表: ['a', 'b', 'c']

print("初始化后的list1:", list1)  # 输出: [1, 2, 3]
print("初始化后的str_list:", str_list)  # 输出: ['a', 'b', 'c']
print("\n===== 基础操作开始 =====")
print()


# ====================== 一、基础属性与简单操作 ======================
# 1.1 获取列表长度（元素个数）
length = len(list1)  # len()是Python内置函数，适用于所有序列（列表、字符串等）
print("1.1 列表长度:", length)  # 输出: 3

# 1.2 获取最大/最小值（仅适用于元素可比较的列表）
max_val = max(list1)  # max()返回列表中的最大值
min_val = min(list1)  # min()返回列表中的最小值
print("1.2 最大值:", max_val, "| 最小值:", min_val)  # 输出: 3 | 1

# 1.3 计算元素总和（仅适用于数值型列表）
total = sum(list1)  # sum()累加所有元素（int/float类型）
print("1.3 元素总和:", total)  # 输出: 6

# 1.4 列表拼接与重复（生成新列表，不修改原列表）
concat_list = list1 + [4, 5]  # 用+拼接两个列表
repeat_list = list1 * 2       # 用*重复列表元素
print("1.4 拼接后的列表:", concat_list)  # 输出: [1, 2, 3, 4, 5]
print("1.4 重复后的列表:", repeat_list)   # 输出: [1, 2, 3, 1, 2, 3]
print()


# ====================== 二、元素访问与切片 ======================
# 列表元素通过索引访问（索引从0开始，负数表示倒数）
# 2.1 单元素访问
first = list1[0]    # 第1个元素（索引0）
last = list1[-1]    # 最后1个元素（索引-1）
print("2.1 第一个元素:", first, "| 最后一个元素:", last)  # 输出: 1 | 3

# 2.2 切片操作（截取子列表，格式：[起始索引:结束索引:步长]）
sub1 = list1[1:3]    # 索引1到2（不包含3）
sub2 = list1[:2]     # 从开头到索引1（等价于[0:2]）
sub3 = list1[::2]    # 步长2（每隔1个元素取一个）
print("2.2 切片sub1:", sub1)  # 输出: [2, 3]
print("2.2 切片sub2:", sub2)  # 输出: [1, 2]
print("2.2 切片sub3:", sub3)  # 输出: [1, 3]
print()


# ====================== 三、元素修改（增删改） ======================
# 3.1 添加元素（修改原列表）
# 3.1.1 末尾添加单个元素
list1.append(4)  # append(x): 末尾添加元素x（仅1个）
print("3.1.1 末尾添加后:", list1)  # 输出: [1, 2, 3, 4]

# 3.1.2 末尾添加多个元素（扩展列表）
list1.extend([5, 6])  # extend(iterable): 末尾添加可迭代对象的所有元素
print("3.1.2 扩展后:", list1)  # 输出: [1, 2, 3, 4, 5, 6]

# 3.1.3 指定位置插入元素
list1.insert(2, 100)  # insert(i, x): 在索引i处插入元素x（i=0为开头）
print("3.1.3 插入后:", list1)  # 输出: [1, 2, 100, 3, 4, 5, 6]

# 3.2 删除元素（修改原列表）
# 注意：若元素不存在，remove()和index()会抛出ValueError！
# 3.2.1 删除首个匹配项
if 100 in list1:  # 先检查元素是否存在（避免报错）
    list1.remove(100)
print("3.2.1 删除100后:", list1)  # 输出: [1, 2, 3, 4, 5, 6]

# 3.2.2 弹出指定位置元素（返回被删除的值）
popped = list1.pop(2)  # pop(i): 删除索引i的元素（默认i=-1，即最后一个）
print("3.2.2 弹出索引2的元素:", popped, "| 剩余列表:", list1)  # 输出: 3 | [1, 2, 4, 5, 6]

# 3.2.3 清空列表
list1.clear()  # 等价于 del list1[:]
print("3.2.3 清空后列表:", list1)  # 输出: []
print()


# ====================== 四、查询与统计 ======================
# 重新初始化列表（含重复元素）
list1 = [1, 2, 3, 3, 4]

# 4.1 查找元素位置（返回首个匹配的索引）
if 3 in list1:  # 先检查存在性
    index = list1.index(3)
    print("4.1 元素3的首个索引:", index)  # 输出: 2

# 4.2 统计元素出现次数
count = list1.count(3)
print("4.2 元素3的出现次数:", count)  # 输出: 2

# 4.3 检查元素是否存在（常用操作）
is_exist = 5 in list1  # 返回布尔值
print("4.3 元素5是否存在:", is_exist)  # 输出: False
print()


# ====================== 五、排序与反转 ======================
# 5.1 排序（两种方式）
# 方式1：sorted()返回新列表（原列表不变）
sorted_new = sorted(list1)  # 默认升序
sorted_rev = sorted(list1, reverse=True)  # 降序
print("5.1 sorted()升序:", sorted_new)  # 输出: [1, 2, 3, 3, 4]
print("5.1 sorted()降序:", sorted_rev)  # 输出: [4, 3, 3, 2, 1]

# 方式2：sort()原地排序（修改原列表）
list1.sort()  # 默认升序
print("5.1 sort()升序后原列表:", list1)  # 输出: [1, 2, 3, 3, 4]
list1.sort(reverse=True)  # 降序
print("5.1 sort()降序后原列表:", list1)  # 输出: [4, 3, 3, 2, 1]

# 5.2 反转列表（原地修改）
list1.reverse()  # 等价于 list1[::-1]（但reverse()直接修改原列表）
print("5.2 反转后列表:", list1)  # 输出: [1, 2, 3, 3, 4]
print()


# ====================== 六、列表复制 ======================
# 注意：直接赋值（list2 = list1）是引用传递，修改list2会影响list1！
# 正确复制方式：
# 方式1：copy()方法（浅拷贝，适用于简单列表）
list_copy1 = list1.copy()
# 方式2：切片复制（list1[:]）
list_copy2 = list1[:]

list_copy1.append(5)  # 修改副本不影响原列表
print("6. 原列表:", list1)        # 输出: [1, 2, 3, 3, 4]
print("6. 复制后的列表:", list_copy1)  # 输出: [1, 2, 3, 3, 4, 5]


# ====================== 总结 ======================
# 列表核心特点：
# - 有序（元素按插入顺序排列）
# - 可变（可增删改元素）
# - 可包含任意类型元素（混合类型：[1, "a", True]）
```
